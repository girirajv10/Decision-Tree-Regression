```python
# Import Packages, Function and Classes

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.tree import plot_tree
```

```python
# Create/Read a dataset

dataset = np.array(
[['Asset Flip', 100, 1000],
['Text Based', 500, 3000],
['Visual Novel', 1500, 5000],
['2D Pixel Art', 3500, 8000],
['2D Vector Art', 5000, 6500],
['Strategy', 6000, 7000],
['First Person Shooter', 8000, 15000],
['Simulator', 9500, 20000],
['Racing', 12000, 21000],
['RPG', 14000, 25000],
['Sandbox', 15500, 27000],
['Open-World', 16500, 30000],
['MMOFPS', 25000, 52000],
['MMORPG', 30000, 80000]
])

print(dataset)
```

```python
# Divide the data into “Attributes” and “labels”

X = dataset[:, 1:2].astype(int)
y = dataset[:, 2].astype(int)
```

```python
# Create a Model and Train it

dtReg=DecisionTreeRegressor()
dtReg.fit(X,y)
```

```python
# Predicting the data

y_predict = dtReg.predict(X)
print(y_predict)
```

```python
# Visualising the Regression results 

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.figure(figsize=(9, 6), dpi=80)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, dtReg.predict(X_grid), color = 'blue')
plt.title('Profit to Production Cost using Decision Tree Regression')
plt.xlabel('Production Cost')
plt.ylabel('Profit')
plt.show()
```

```python
# Visualising the graph without the use of graphvizplt.figure(figsize = (20,20))

plt.figure(figsize=(16,12))
dec_tree = plot_tree(decision_tree=dtReg, 
                     feature_names = ['Production Cost','Profit'], 
                     filled = True ,
                     precision = 4, 
                     rounded = True)
plt.show()
```
